```python
# i realized that after i concatenated the both data set ticker_symbol and company_name had null values 
#i will be using the merge to solve the values mistakes 

Data = pd.merge(Company_data, Company_name, on="ticker_symbol", how="left") # on -Ensures the join is done correctly on this column
                                                                             # how =left - it Keeps all rows from Company_data and adds corresponding
                                                                                #company_name.
```

```python
# Creating  a dictionary from available ticker-symbol mappings
ticker_mapping = Company_name.set_index("ticker_symbol")["company_name"].to_dict()
```

```python
# Fill missing values in Data
Data["company_name"] = Data["ticker_symbol"].map(ticker_mapping) #this is where the magic is i filled it up using the mapping value to enable it correspond 
                                                                 # with our previous code which states is on=ticker_symbol and how=left
```
